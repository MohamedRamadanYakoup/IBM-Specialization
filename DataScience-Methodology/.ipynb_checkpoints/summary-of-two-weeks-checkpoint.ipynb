### The Structure of Powerful Decision Tree:

* Decision trees are built using recursive partitioning to classify the data.
 - **Recursive partitioning creates a decision tree that strives to correctly classify members of the population by splitting it into sub-populations based on several dichotomous independent variables.**

<br>

* When partitioning the data, decision trees use the most predictive feature (ingredient in this case) to split the data.
* **Predictiveness** is based on decrease in entropy - gain in information, or *impurity*.

```python
df_counts = df['cuisine'].value_counts()
df_indexs = df_counts > 50

df_target = list(np.array(df_counts.index.values)[np.array(df_indexs)])
```

```python
before = df.shape[0]
print("Data Rows Before removing the unneeded data: {}".format(before))


df=df.loc[df['cuisine'].isin(df_target)]
after = df.shape[0]
print("Data Rows After removing the unneeded data: {}".format(df.shape[0]))
```

```python
df = df.replace("Yes", 1)
df = df.replace("No", 0)
df.head()
```

```python
target_df = df.loc[(df['rice']==1) & (df['soy_sauce']==1) & (df['wasabi']==1) & df['seaweed']]
target_df
```

```python
ing = df.iloc[:,1:].sum(axis=0).to_frame().reset_index()
ing.rename(columns={"index":"ingradient", 0:"count"}, inplace=True)
ing
```

```python
ing.sort_values(['count'], ascending=False, inplace=True)
ing.reset_index(inplace=True, drop=True)
ing
```

```python
cuisines = df.groupby("cuisine").mean()
cuisines.head()
```

```python
max_4 = 4

def print_top_ing(data):
    print(data.name.upper())
    print("-"*len(data.name))
    row_sorted = data.sort_values(ascending=False)*100
    top_ing = list(row_sorted.index.values)[0:max_4]
    row_sorted = list(row_sorted)[0:max_4]
    
    for ind, ing in enumerate(top_ing):
        print("%s (%d%%) |"%(ing, row_sorted[ind]), end=' ')
    print("\n")
    
cuisines_profiles = cuisines.apply(print_top_ing, axis=1)
```

## Data Modeling

---

```python
%matplotlib inline
from sklearn import tree
from sklearn.metrics import accuracy_score, confusion_matrix

import matplotlib.pyplot as plt

!conda install python-graphviz --yes
import graphviz

from sklearn.tree import export_graphviz

import itertools
```

```python
asian_df = df[df.cuisine.isin(["korean", "japanese", "chinese", "thai", "indian"])]
cuisines = asian_df['cuisine']
ingradients = asian_df.iloc[:,1:]

dec_tree = tree.DecisionTreeClassifier(max_depth=3)
dec_tree.fit(ingradients, cuisines)
```

```python
export_graphviz(dec_tree, feature_names=list(ingradients.columns.values), 
                out_file="dec_tree.dot", class_names=np.unique(cuisines), 
               filled=True,
                node_ids=True,
                special_characters=True,
                impurity=False,
                label="all",
                leaves_parallel=False)

with open("dec_tree.dot") as dec_tree_img:
    dec_tree_graph = dec_tree_img.read()
    
graphviz.Source(dec_tree_graph)
```

## Model Evaluation

---

```python
bamboo = df[df.cuisine.isin(["korean", "japanese", "chinese", "thai", "indian"])]
```

```python
bamboo['cuisine'].value_counts()
```

```python
np.random.seed(1234)

test_bamboo = bamboo.groupby('cuisine', group_keys=False).apply(lambda x: x.sample(30))
test_ing = test_bamboo.iloc[:, 1:]
test_cuisine = test_bamboo['cuisine']

test_bamboo['cuisine'].value_counts()
```

```python
bamboo_index_test = bamboo.index.isin(test_bamboo.index)
bamboo_train = bamboo[~bamboo_index_test]
train_ing = bamboo_train.iloc[:,1:]
train_cuisine = bamboo_train['cuisine']

bamboo_train['cuisine'].value_counts()
```

```python
train_tree = tree.DecisionTreeClassifier(max_depth=15)
train_tree.fit(train_ing, train_cuisine)
```

```python
export_graphviz(train_tree,
                feature_names=list(train_ing.columns.values),
                out_file="bamboo_train_tree.dot",
                class_names=np.unique(train_cuisine),
                filled=True,
                node_ids=True,
                special_characters=True,
                impurity=False,
                label="all",
                leaves_parallel=False)

with open("bamboo_train_tree.dot") as bamboo_train_tree_image:
    bamboo_train_tree_graph = bamboo_train_tree_image.read()
graphviz.Source(bamboo_train_tree_graph)
```

```python
pred_cuisine = train_tree.predict(test_ing)
```

```python
test_cuisines = np.unique(test_cuisine)
bamboo_confusion_matrix = confusion_matrix(test_cuisine, pred_cuisine, test_cuisines)
title = 'Bamboo Confusion Matrix'
cmap = plt.cm.Blues

plt.figure(figsize=(8, 6))
bamboo_confusion_matrix = (
    bamboo_confusion_matrix.astype('float') / bamboo_confusion_matrix.sum(axis=1)[:, np.newaxis]
    ) * 100

plt.imshow(bamboo_confusion_matrix, interpolation='nearest', cmap=cmap)
plt.title(title)
plt.colorbar()
tick_marks = np.arange(len(test_cuisines))
plt.xticks(tick_marks, test_cuisines)
plt.yticks(tick_marks, test_cuisines)

fmt = '.2f'
thresh = bamboo_confusion_matrix.max() / 2.
for i, j in itertools.product(range(bamboo_confusion_matrix.shape[0]), range(bamboo_confusion_matrix.shape[1])):
    plt.text(j, i, format(bamboo_confusion_matrix[i, j], fmt),
             horizontalalignment="center",
             color="white" if bamboo_confusion_matrix[i, j] > thresh else "black")

plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')

plt.show()
```
